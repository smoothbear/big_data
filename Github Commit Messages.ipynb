```python
import requests
import time

baseURL = 'https://api.github.com/repos/{owner}/{repo}/commits?per_page=100'

headers = {'User-Agent' : 'Mozilla/4.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}

repositories = {
    "vue": {
        "owner": "vuejs",
        "repo": "vue",
        "messages": []
    },
    "linux": {
        "owner": "torvalds",
        "repo": "linux",
        "messages": []
    },
    "react": {
        "owner": "facebook",
        "repo": "react",
        "messages": []
    },
    "flutter": {
        "owner": "flutter",
        "repo": "flutter",
        "messages": []
    },
    "tensorflow": {
        "owner": "tensorflow",
        "repo": "tensorflow",
        "messages": []
    },
    "bootstrap": {
        "owner": "twbs",
        "repo": "bootstrap",
        "messages": []
    }
}

for key, value in repositories.items():
    changedOwner = baseURL.replace("{owner}", value.get("owner"))
    URL = changedOwner.replace("{repo}", value.get("repo"))

    for page in range(1, 5):
        response = requests.get(URL + "&page=" + str(page), headers=headers)
        if response.status_code != 200:
            break

        for commit in response.json():
            value.get("messages").append((commit.get('commit').get('message')))



repositories
```

```python
import re

patterns = {
    "vue": {
        "pattern_1": [],
        "pattern_2": [],
        "pattern_3": [],
        "pattern_4": [],
        "other_pattern": []
    },
    "linux": {
        "pattern_1": [],
        "pattern_2": [],
        "pattern_3": [],
        "pattern_4": [],
        "other_pattern": []
    },
    "react": {
        "pattern_1": [],
        "pattern_2": [],
        "pattern_3": [],
        "pattern_4": [],
        "other_pattern": []
    },
    "flutter": {
        "pattern_1": [],
        "pattern_2": [],
        "pattern_3": [],
        "pattern_4": [],
        "other_pattern": []
    },
    "tensorflow": {
        "pattern_1": [],
        "pattern_2": [],
        "pattern_3": [],
        "pattern_4": [],
        "other_pattern": []
    },
    "bootstrap": {
        "pattern_1": [],
        "pattern_2": [],
        "pattern_3": [],
        "pattern_4": [],
        "other_pattern": []
    },
    "all": {
        "pattern_1": [],
        "pattern_2": [],
        "pattern_3": [],
        "pattern_4": [],
        "other_pattern": []
    }
}

pattern_1_regex = re.compile('(?i)([+[A-Z])+\w+]') # [ADD] prefix
pattern_2_regex = re.compile('(?i)([A-Z])\w+:') # Add: prefix
pattern_3_regex = re.compile('(?i)([+[A-Z])+\w+\([^)]*\):')
pattern_4_regex = re.compile('(?i)([A-Z])\w')

for key, value in repositories.items():
    for message in value.get("messages"):
        commitTitle = message.split("\n")[0]

        if commitTitle.startswith("Merge"):
            continue

        if pattern_1_regex.match(commitTitle):
            patterns.get(key).get("pattern_1").append(commitTitle)
            patterns.get("all").get("pattern_1").append(commitTitle)

        elif pattern_2_regex.match(commitTitle):
            patterns.get(key).get("pattern_2").append(commitTitle)
            patterns.get("all").get("pattern_2").append(commitTitle)
        
        elif pattern_3_regex.match(commitTitle):
            patterns.get(key).get("pattern_3").append(commitTitle)
            patterns.get("all").get("pattern_3").append(commitTitle)

        elif pattern_4_regex.match(commitTitle):
            patterns.get(key).get("pattern_4").append(commitTitle)
            patterns.get("all").get("pattern_4").append(commitTitle)

        else:
            patterns.get(key).get("other_pattern").append(commitTitle)
            patterns.get("all").get("other_pattern").append(commitTitle)

print("-------------------------------------------------------------------------\n\nresult: ")
print("pattern 1([ADD] prefix) : ", str(len(patterns.get("all").get("pattern_1"))))
print("pattern 2(Add: prefix) : ", str(len(patterns.get("all").get("pattern_2"))))
print("pattern 3(add([what fixed]): prefix) : ", str(len(patterns.get("all").get("pattern_3"))))
print("pattern 4(Add prefix) : ", str(len(patterns.get("all").get("pattern_4"))))
print("other_pattern : ", str(len(patterns.get("all").get("other_pattern"))))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.animation as animation
import seaborn as sns
```

```python
plt.rc('font', family='NanumBarunGothic')
%matplotlib inline
```

```python
def draw_barchart(name):
    data = patterns.get(name) 
    count1 = len(data.get("pattern_1"))
    count2 = len(data.get("pattern_2"))
    count3 = len(data.get("pattern_3"))
    count4 = len(data.get("pattern_4"))
    count5 = len(data.get("other_pattern"))

    x = np.arange(5)
    colors = sns.color_palette('hls', 5)

    plt.title(f"{name} commit prefixes", fontsize=14)
    plt.xlabel('prefixes', fontsize=12)
    plt.ylabel('Number of commit messages', fontsize=12)
    plt.bar(x, [count1, count2, count3, count4, count5], tick_label=["[Add]", "Add:", "add(changed):", "Add ", "Others"], color=colors, edgecolor=colors, alpha=0.7, linewidth=4)

draw_barchart("react")
```
